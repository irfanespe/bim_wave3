Buat code untuk menentukan jenis segitiga dengan ketentuan berikut
1. panjang sisi segitiga adalah a, b, dan c.
2. ketiga panjang sisi segitiga merupakan variable yang bisa diubah-ubah value nya
3. jika panjang a, b, dan c sama Maka code mengeluarkan tulisan `segitiga sama sisi`
4. jika panjang a dan b sama, maka code mengeluarkan tulisan `segitiga sama kaki`
