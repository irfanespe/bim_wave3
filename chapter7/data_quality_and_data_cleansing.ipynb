### b.2.i filling with unknown / impossible value

```python
# filling dengan unknown / value yang ga mungkin ada -> case ini kita isi `invalid`
# untuk mengisi yang null , template -> nama_var['column'].fillna('value')
# cara ini sering dipake di BI untuk maintain datawarehouse

df_copy1['size'].fillna("Invalid Size", inplace = True)
```

```python
df_copy1.head()
```

```python
# check null data
df_copy1[df_copy1['size'].isna()]
```

```python
# check value
df_copy1[df_copy1['size'] == 'Invalid Size']
```

### b.2.ii. Filling with the most occurences value

```python
# diisi dengan value terbanyak pada suatu kolom
df_copy2 = df.copy()
```

```python
df_copy2['size'].value_counts()
```

```python
# fill with 'XL'
df_copy2.fillna("XL", inplace = True)
```

```python
# checking data
df_copy2[df_copy2['size'].isna()]
```

Jika column yang null itu merupakan angka, ada beberapa opsi filling alinnya
1. rata2 value pada kolom -> nama_var['column'].mean() -> .fillna(rata-rata)
2. median value pada kolom -> nama_var['column'].median() -> .fillna(median)

```python
# contoh mencari rata2
df['unitPrice'].mean()
```

```python
# opsi 1
# df['unitPrice'].fillna(510.7)

# opsi 2
df['unitPrice'].fillna(df['unitPrice'].mean())
```
