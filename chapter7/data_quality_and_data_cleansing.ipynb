## 1. Data quality 
a. uniqueness (pelanggaran : duplikat) -> kita cek bahwa tidak ada data yang duplikat. contoh : primary key muncul beberapa kali \
b. completeness -> datanya ada yang null atau tidak \
c. accuracy -> value nya itu dalam range yang diperbolehkan atau value real \
d. consistancy -> penulisan nya sama atau tidak / jenisnya seragam atau tidak

### a.uniqueness

```python
# cek duplikat
# template -> df.duplicated() -> cek duplication
df.duplicated()
```

```python
# jumlah row yang duplicate
sum(df.duplicated())
```

```python
# memunculkan data yang duplicate
# template -> nama_va[nama_var.duplicated()]
df[df.duplicated()]
```

Terdapat duplicate data sejumlah 1 data, dengan order_id = 201

```python
df[['order_id', 'order_date']].groupby(by = 'order_id').count()
# order_id bukan primary key
```

```python
# contoh checking duplicate untuk primary key
df[df.duplicated('order_id')]
```

### b. completeness

```python
# template -> nama_var.isna().any()
df.isna().any()
# ada value null pada kolom size, dapat dilihat dari hasil balikan True
```

```python
# check null values in a column
# template -> nama_var['column'].isna()
df['size'].isna()
```

```python
# count total null rows
sum(df['size'].isna())
```

```python
df[df['size'].isna()]
```

Data null terdapat pada kolom `size` sebanyak 36 kolom
